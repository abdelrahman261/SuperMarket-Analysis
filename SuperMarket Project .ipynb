```python
df['Months']=df['Time'].dt.month
```

##### 1- what is the most expensive product on the supermarket and most product line sells



```python
x=df.groupby('Product line').max()['Total']
x.max()
print('Fashion accessories is the most products sells=',x.max())
```

```python
sns.boxenplot(data=df,x='Total',y='Product line',width=.8, hue='City',palette="Set3")
plt.legend( loc ="upper right")
```

#### 2- what is the highest rating in all products



```python
df.groupby('Product line').max()['Rating']
print('Electronic accessories is the highest products rating=',df.groupby('Product line').max()['Rating']

.max())   
```

```python
sns.displot(data=df,x='Rating')
```

#### 3- what is the city have highest products sells and its brach



```python
B=df.groupby('Total').max()['Branch']
B.max()
C=df.groupby('Total').max()['City']
C.max()
print(f'The city that had a higest products sells is: Naypyitaw \n its branch : {B.max()}')
```

```python
pv=pd.pivot_table(index='City',columns='Branch',values='Total',data=df)
pv
```

```python
sns.heatmap(pv)
```

```python
sns.boxplot(data=df, x='City',y='Total', hue='Branch')
plt.legend( loc ="upper center")
```
